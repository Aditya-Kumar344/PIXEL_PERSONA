#paragraph aabout dataset:

the dataset calculates loan_status(if the person should be approved a loan or not, 1 for yes, 0 for no) from the given information about the person asking for loan(like gender, age, education, income, home ownership,credit score, etc) and the person approving the loan(the employee's experience),loan amount.

dropping unnecessory columns and changing column as per requirement

```python
df.drop(['person_gender','person_education','person_home_ownership','loan_intent'],inplace=True,axis=1)
df['previous_loan_defaults_on_file'] = df['previous_loan_defaults_on_file'].map({'Yes': 1, 'No': 0})
```

dropping any NaN values rows

```python
df.dropna(inplace=True)
```

splitting model into two parts

```python
X = df.drop(columns=['loan_status'])
y=df['loan_status']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.8,random_state=10)
```

using imported logistic reg model

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(tol=0.01, max_iter=1000)
model.fit(X_train,y_train)
```

bulding my own Logistic reg model

```python
import numpy as np
import warnings
warnings.filterwarnings("ignore")
class my_Logistic_Regression():

  def __init__(self, learning_rate, iterations):

    self.learning_rate = learning_rate
    self.iterations = iterations

  def fit(self, X, Y):

    self.m, self.n = X.shape

    self.w = np.zeros(self.n)

    self.b = 0

    self.X = X

    self.Y = Y

    for i in range(self.iterations):
      self.update_weights()

  def update_weights(self):

    Y_hat = 1 / (1 + np.exp( - (self.X.dot(self.w) + self.b ) ))

    dw = (1/self.m)*np.dot(self.X.T, (Y_hat - self.Y))

    db = (1/self.m)*np.sum(Y_hat - self.Y)

    self.w = self.w - self.learning_rate * dw

    self.b = self.b - self.learning_rate * db

  def predict(self, X):

    Y_pred = 1 / (1 + np.exp( - (X.dot(self.w) + self.b ) ))
    Y_pred = np.where( Y_pred > 0.5, 1, 0)
    return Y_pred
```

```python
my_model=my_Logistic_Regression(learning_rate=0.01,iterations=1000)
my_model.fit(X_train,y_train)
```

accuracy testing

```python
from sklearn.metrics import accuracy_score
y_pred=model.predict(X_test)
imported_accuracy=accuracy_score(y_test, y_pred)
print("accuracy_score of imported model : ",imported_accuracy*100)
y_pred=my_model.predict(X_test)
my_model_accuracy=accuracy_score(y_test, y_pred)
print("accuracy_score of my model : ",my_model_accuracy*100)
```
